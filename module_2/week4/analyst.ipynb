```python
import pandas as pd
data = pd.read_csv("advertising.csv")
def correlation(x, y):
    x_mean = x.mean()
    y_mean = y.mean()
    x_diff = x - x_mean
    y_diff = y - y_mean
    numerator = (x_diff * y_diff).sum()
    denominator = ((x_diff ** 2).sum() * (y_diff ** 2).sum()) ** 0.5
    return numerator / denominator
x = data['TV']
y = data['Radio']
corr_xy = correlation(x, y)
print(f"Correlation between TV and Radio: {round(corr_xy, 2)}")
```

```python
feature = ['TV', 'Radio', 'Newspaper']
for feature_1 in feature:
    for feature_2 in feature:
        correlation_value = correlation(data[feature_1], data[feature_2])
        print(f"Correlation between {feature_1} and {feature_2}: {round(correlation_value, 2)}")
```
